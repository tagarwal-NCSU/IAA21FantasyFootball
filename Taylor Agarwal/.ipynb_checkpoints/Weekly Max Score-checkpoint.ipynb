```python
import pandas as pd
import numpy as np

path = "../IAA League Data/MATCHUP_DATA_2021.xlsx"
df = pd.read_excel(path)
df
```

```python
teams = df["Team"].unique()
slots = ["QB", "RB", "RB", "WR", "WR", "TE", "RB/WR/TE", "D/ST", "K"]

for week in range(1, 18):
    week_df = df[df["Week"] == week]
    for team in teams:
        team_df = week_df[week_df["Team"] == team]
        team_df = team_df.sort_values("Points", ascending = False)
        first = True
        for slot in slots:
            top_player = team_df[team_df["Eligible Slots"].str.split(", ").contains(slot)].nlargest(1,'Points')
            if first:
                best_df = top_player
                first = False
            else:
                best_df.append(top_player)
            i = top_player.index
            team_df = team_df.drop(i)
        print(best_df)
```
