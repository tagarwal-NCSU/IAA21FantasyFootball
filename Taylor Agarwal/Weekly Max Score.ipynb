```python
import pandas as pd
import numpy as np

path = "../IAA League Data/MATCHUP_DATA_2021.xlsx"
df = pd.read_excel(path)
df["Eligible Slots"] = " " + df["Eligible Slots"] + ","
df
```

```python
teams = df["Team"].unique()
slots = ["QB", "RB", "RB", "WR", "WR", "TE", "RB/WR/TE", "D/ST", "K"]

first_df = True
for week in range(1, 18):
    week_df = df[df["Week"] == week]
    for team in teams:
        team_df = week_df[week_df["Team"] == team]
        team_df = team_df.sort_values("Points", ascending = False)
        first = True
        for slot in slots:
            top_player = team_df[team_df["Eligible Slots"].str.contains(" "+slot+",")].nlargest(1,'Points')
            if first:
                best_df = top_player
                first = False
            else:
                best_df = best_df.append(top_player)
            i = top_player.index
            team_df = team_df.drop(i)
        if first_df:
            final_df = best_df
            first_df = False
        else:
            final_df = final_df.append(best_df)
final_df
```

```python
scoring_df = df[~df["Slot"].isin(["BE", "IR"])]
actual_scores = scoring_df.groupby(["Week", "MatchupId", "Team"]).sum().reset_index()
best_scores = final_df.groupby(["Week", "MatchupId", "Team"]).sum().reset_index()
```

```python
actual_scores.to_excel("../IAA League Data/ACTUAL_SCORES_2021.xlsx", index = False)
best_scores.to_excel("../IAA League Data/BEST_SCORES_2021.xlsx", index = False)
```
