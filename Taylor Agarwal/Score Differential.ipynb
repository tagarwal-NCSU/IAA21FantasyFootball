```python
import pandas as pd

path = "C:/Users/taywo/Documents/Git Repos/IAA21FantasyFootball/Andrew's League Data/MATCHUP_DATA_2020.csv"

df = pd.read_csv(path)
df
```

```python
num_weeks = int(df.Week.max())
teams = list(df.Team.unique())

max_weekly_scores = dict()
roster = ["QB", "RB", "WR", "RB/WR", "RB/WR", "TE", "D/ST", "K"]
df["Eligible Slots"] = df["Eligible Slots"].str.replace("RB/WR/TE", "")
df["Eligible Slots"] = df["Eligible Slots"].str.replace("WR/TE", "")
df["Eligible Slots"] = df["Eligible Slots"].str.replace("RB/WR", "")

for team in teams:
    team_df = df[df.Team.isin([team])]
    for week in range(1, num_weeks+1):
        team_week_df = team_df[team_df.Week.isin([week])]
        best_team = {'positions': [],
                     'players': [],
                     'points': []
                    }
        for position in roster:
            if position != "RB/WR":
                position_df = team_week_df[team_week_df["Eligible Slots"].str.contains(position)]
            else:
                position_df = team_week_df[team_week_df["Eligible Slots"].str.contains("RB") |
                                           team_week_df["Eligible Slots"].str.contains("WR")]
            max_idx = position_df.Points.idxmax()
            best_team['positions'].append(position)
            best_team['players'].append(position_df.iloc[max_idx, 4])
            best_team['points'].append(position_df.iloc[max_idx, 6])
            team_week_df = team_week_df.drop(index=max_idx)
        print(best_team)
```
